```python
import os
os.chdir('C:\\Users\\jenat\\OneDrive\\Documents\\160\\shapefiles\\census_tracts\\arizona')
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib.cm
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
import geopandas as gp
import shapefile
import pandas as pd
import numpy as np

az = gp.GeoDataFrame.from_file('new.shp')
```

```python
y = az.N_LAT.values
x = az.N_LON.values
```

```python
fig, ax1 = plt.subplots(figsize=(18, 45))
m = Basemap(resolution='i',projection='merc',llcrnrlat=31.165, urcrnrlat=36.991, llcrnrlon=-114.822, urcrnrlon=-109.219)
m.drawcounties(color='purple')
m.fillcontinents(color='#f2f2f2',lake_color='#46bcec')
x1,y1 = m(x, y)
m.scatter(x1, y1, c = 'r',marker = "o", alpha = 1.0, zorder=10)
m.drawmapboundary(fill_color='#46bcec')
  
m.readshapefile('new','az',linewidth = 2, color = 'black', drawbounds=True)
plt.title('Arizona Health Centres', fontsize = 25)

red_dot = mlines.Line2D([], [], color='red', marker='o',markersize=9, label='Store Locations')
plt.legend(handles=[red_dot])


plt.show()
```
