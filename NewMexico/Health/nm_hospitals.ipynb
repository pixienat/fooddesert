```python
import os
os.chdir('C:\\Users\\jenat\\OneDrive\\Documents\\160\\shapefiles\\census_tracts\\arizona')
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib.cm
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
import geopandas as gp
import shapefile
import pandas as pd
import numpy as np

nm = gp.GeoDataFrame.from_file('nm_health.shp')
```

```python
y = nm.GC_LAT.values
x = nm.GC_LONG.values
```

```python
fig, ax1 = plt.subplots(figsize=(18, 45))
m = Basemap(resolution='i',projection='merc',llcrnrlat=31.310, urcrnrlat=37.038, llcrnrlon=-109.101, urcrnrlon=-103.023)
m.drawcounties(color='purple')
m.fillcontinents(color='#f2f2f2',lake_color='#46bcec')
x1,y1 = m(x, y)
m.scatter(x1, y1, c = 'b',marker = "o", alpha = 1.0, zorder=10)
m.drawmapboundary(fill_color='#46bcec')
  
m.readshapefile('nm_health','nm',linewidth = 2, color = 'black', drawbounds=True)
plt.title('New Mexico Hospitals', fontsize = 25)

blue_dot = mlines.Line2D([], [], color='blue', marker='o',markersize=9, label='Hospital Locations')
plt.legend(handles=[blue_dot])



plt.show()
```
